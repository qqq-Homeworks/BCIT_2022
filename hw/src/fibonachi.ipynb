```python
import requests
```

```python
import matplotlib.pyplot as plt
```

```python
n = 100
```

```python
r = requests.get('http://127.0.0.1:5000/'+n.__str__())
```

```python
numbers = list(r.json())
```

```python
plt.plot(numbers)
plt.ylabel('fibonachi for ' + n.__str__()+ '  numbers')
plt.show()
```
